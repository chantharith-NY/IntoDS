```python
# encode categorical data into numeric values

# Step 1: Assign consecutive numbers
color_mapping = {category: idx+1 for idx, category in enumerate(cleaned_data["Color"].unique())}
source_mapping = {category: idx+1 for idx, category in enumerate(cleaned_data["Source"].unique())}

# Step 2: Apply the mapping
cleaned_data["Color"] = cleaned_data["Color"].map(color_mapping)
cleaned_data["Source"] = cleaned_data["Source"].map(source_mapping)

# Check the result
print(cleaned_data[["Source","Color"]])
```

Check outlier

```python

# Select numeric columns
numeric_cols = ['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead', 'Zinc', 
                'Turbidity', 'Fluoride', 'Copper', 'Sulfate', 
                'Conductivity', 'Chlorine', 'Manganese', 
                'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']

# Function to calculate outliers
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Create a summary report
summary = []

# Detect outliers for each numeric column
for col in numeric_cols:
    outlier_data, lower, upper = detect_outliers_iqr(df, col)
    count_outliers = len(outlier_data)
    if count_outliers > 0:
        summary.append({
            "Column": col,
            "Outlier Count": count_outliers,
            "Lower Bound": lower,
            "Upper Bound": upper
        })
    else:
        print(f"\nColumn: {col} - No outliers detected.")

# Convert summary to DataFrame for easier viewing
summary_df = pd.DataFrame(summary)
summary_df
```

```python
# PLot the box plot to see the outlier clearly

import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for each numeric column
plt.figure(figsize=(15, 10))
sns.boxplot(data=cleaned_data[numeric_cols], orient='h')
plt.title("Boxplots of Numeric Columns")
plt.grid(True)
plt.show()
```

+ Base on the data and graph, we can see that there are tons of outlier in our dataset. 
  - This outlier maybe due to values errors or natural anomalies

+ To solve this outlier we should use some engineer technique (log transform, square root, clip or clap...) instead of remove all of it

Before we can decide which of this feature need to use any engineer technique or remove we need to find the skewness of it.

```python

import matplotlib.pyplot as plt
from scipy.stats import skew

# Define numeric columns
numeric_cols = ['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead', 'Zinc', 
                'Turbidity', 'Fluoride', 'Copper', 'Sulfate', 
                'Conductivity', 'Chlorine', 'Manganese', 
                'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']

# Check skewness for each feature
skewness_values = cleaned_data[numeric_cols].skew()

# Function to apply transformations based on skewness
def apply_transformations(data, skewness_values):
    for feature in numeric_cols:
        skew_value = skewness_values[feature]
        
        if skew_value > 1:
            # Highly skewed, apply log transformation
            print(f"Feature {feature} is highly skewed ({skew_value:.2f}), applying log transformation.")
            data[feature] = np.log1p(data[feature])  # log(x + 1) to handle zero values
        elif skew_value < -1:
            # Highly negatively skewed, apply log transformation (if the data allows)
            print(f"Feature {feature} is highly negatively skewed ({skew_value:.2f}), applying log transformation.")
            data[feature] = np.log1p(data[feature])  # log(x + 1) to handle zero values
        elif 0.5 < abs(skew_value) <= 1:
            # Mildly skewed, apply square root transformation
            print(f"Feature {feature} is mildly skewed ({skew_value:.2f}), applying square root transformation.")
            data[feature] = np.sqrt(data[feature])
        elif -0.5 <= skew_value <= 0.5:
            # Symmetric distribution, no transformation needed
            print(f"Feature {feature} is symmetrically distributed ({skew_value:.2f}), no transformation applied.")
            continue
        
        # Optionally apply clipping if needed 
        # if feature in ['Lead', 'Zinc', 'Manganese']: 
        #     upper_limit = data[feature].quantile(0.95)
        #     lower_limit = data[feature].quantile(0.05)
        #     print(f"Feature {feature}: Clipping outliers to 95th percentile.")
        #     data[feature] = np.clip(data[feature], lower_limit, upper_limit)
        
    return data

# Apply the transformations
transformed_data = apply_transformations(cleaned_data.copy(), skewness_values)

# Optionally, you can check the new skewness after transformations
new_skewness = transformed_data[numeric_cols].skew()
print("\nNew Skewness after transformations:")
print(new_skewness)

# Visualize original vs. transformed data
fig, ax = plt.subplots(2, 1, figsize=(20, 15))  # Adjust subplot grid to be 2 rows, 1 column

# Original Data Distribution
cleaned_data[numeric_cols].hist(bins=50, ax=ax[0])
ax[0].set_title("Original Data Distribution")
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Frequency')

# Transformed Data Distribution
transformed_data[numeric_cols].hist(bins=50, ax=ax[1])
ax[1].set_title("Transformed Data Distribution")
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

+ From the updated skewness values after applying transformations, some features are still highly skewed
    - Features with High Skewness (Post-Transformation):
        - Iron: 4.88
        - Lead: 35.23
        - Turbidity: 1.97
        - Copper: 1.32
        - Manganese: 5.16
        - Despite log transformations, these features remain highly skewed.

+ For features still highly skewed after log transformation, consider further adjustments:
    - Apply Box-Cox transformation
    - clip the extreme value of highly skew


```python
from scipy.stats import boxcox

# Using Box Cox method
transformed_data['Iron'], _ = boxcox(transformed_data['Iron'] + 1)  
transformed_data['Turbidity'], _ = boxcox(transformed_data['Turbidity'] + 1)  
transformed_data['Copper'], _ = boxcox(transformed_data['Copper'] + 1) 
transformed_data['Manganese'], _ = boxcox(transformed_data['Manganese'] + 1) 

# Clipping 'Lead' at its 95th percentile
upper_limit = transformed_data['Lead'].quantile(0.95)
transformed_data['Lead'] = np.clip(transformed_data['Lead'], None, upper_limit)

#  Check new skewness
new_skewness = transformed_data[["Iron","Turbidity","Copper","Manganese","Lead"]].skew()
print("\nNew Skewness after transformations:")
print(new_skewness)
```
